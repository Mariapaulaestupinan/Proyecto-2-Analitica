A partir de los datos que tenemos, queremos identificar relaciones entre las columnas que contienen datos completos y aquellas que tienen datos faltantes, con el fin de poder estimar y completar los valores ausentes en lugar de eliminar las filas incompletas.

```python
df_copia_datos = df.copy()
```

```python
import pandas as pd

# Crear los rangos de puntaje
rango_puntaje_global = [0, 20, 40, 60, 80, 100]
rango_labels = ['0-20', '20-40', '40-60', '60-80', '80-100']
df['rango_puntaje'] = pd.cut(df['punt_ingles'], bins=rango_puntaje_global, labels=rango_labels, right=False)

# Tabla cruzada: cuenta cuántos "S" y "N" hay por cada rango
tabla_cruzada = pd.crosstab(df['rango_puntaje'], df['cole_bilingue'])

# Convertimos a porcentaje sobre el total por fila (por rango)
tabla_porcentajes_puntaje_colegio_bilingue = tabla_cruzada.div(tabla_cruzada.sum(axis=1), axis=0).round(4) * 100

# Mostrar la tabla final
print(tabla_porcentajes_puntaje_colegio_bilingue)
```

```python
#Al DataFrame se le agrega una columna llamada rango_puntaje, que asigna un rango basado en el puntaje de inglés.
rango_puntaje_global_ajustado = [-0.01] + rango_puntaje_global[1:]

df_copia_datos['rango_puntaje'] = pd.cut(
    df_copia_datos['punt_ingles'],
    bins=rango_puntaje_global_ajustado,
    labels=rango_labels,
    right=True  
)
```

```python
def asignar_cole_bilingue_segun_puntaje(df, tabla_porcentajes, col_rango='rango_puntaje', col_obj='cole_bilingue', margen=5):
    # Crear una copia del DataFrame original para no modificarlo directamente
    df = df.copy()

    # Obtener la lista de categorías (valores posibles en la columna objetivo)
    categorias = tabla_porcentajes.columns.tolist()  
    
    # Iterar sobre cada rango de puntaje definido en el índice de la tabla de porcentajes
    for rango in tabla_porcentajes.index:
        # Guardar en un diccionario el porcentaje objetivo para cada categoría en este rango
        porcentaje_obj = tabla_porcentajes.loc[rango].to_dict()

        # Filtrar el DataFrame para quedarnos solo con las filas que pertenecen a este rango
        mask_rango = df[col_rango] == rango
        df_rango = df.loc[mask_rango]

        # Contar cuántas veces aparece cada categoría en la columna objetivo dentro de este rango
        conteo_actual = df_rango[col_obj].value_counts(dropna=True).to_dict()

        # Asegurar que todas las categorías estén presentes en el conteo, aunque sea con valor 0
        for cat in categorias:
            if cat not in conteo_actual:
                conteo_actual[cat] = 0

        # Obtener los índices de las filas donde la columna objetivo está vacía (NaN) para asignarles categoría
        idx_nan = df_rango[df_rango[col_obj].isna()].index.tolist()

        # Para cada fila con valor NaN en la columna objetivo, intentar asignar una categoría adecuada
        for idx in idx_nan:
            asignado = False

            # Ordenar las categorías según el porcentaje objetivo (de mayor a menor)
            orden_categorias = sorted(categorias, key=lambda x: porcentaje_obj[x], reverse=True)

            # Probar asignar cada categoría en orden hasta encontrar una que no supere el margen permitido de desviación
            for cat in orden_categorias:
                # Crear una copia temporal del conteo actual y sumar uno a la categoría que se está evaluando
                conteo_temp = conteo_actual.copy()
                conteo_temp[cat] += 1

                # Calcular el nuevo porcentaje temporal de cada categoría después de la asignación hipotética
                total_temp = sum(conteo_temp.values())
                porcentaje_temp = {k: (v / total_temp) * 100 for k, v in conteo_temp.items()}

                # Verificar si la desviación entre los porcentajes actuales y objetivos está dentro del margen permitido para todas las categorías
                desviacion_ok = all(abs(porcentaje_temp.get(k, 0) - porcentaje_obj[k]) <= margen for k in porcentaje_obj)

                # Si la desviación está dentro del margen, asignar esta categoría y actualizar el conteo actual
                if desviacion_ok:
                    df.at[idx, col_obj] = cat
                    conteo_actual = conteo_temp
                    asignado = True
                    break  # Salir del ciclo de categorías, porque ya se asignó una categoría válida

            # Si no se pudo asignar ninguna categoría que cumpla con el margen, asignar la categoría con mayor porcentaje objetivo por defecto
            if not asignado:
                cat = orden_categorias[0]
                df.at[idx, col_obj] = cat
                conteo_actual[cat] += 1

    # Retornar el DataFrame modificado con las asignaciones hechas
    return df
```

```python
#Apliar la función
df_final = asignar_cole_bilingue_segun_puntaje(df_copia_datos, tabla_porcentajes_puntaje_colegio_bilingue)
df_final
```

```python
#Validar en el DataFrame final la proporción de cada valor de la columna cole_bilingue dentro de cada rango de puntaje.
tabla_porcentajes_puntaje_colegio_bilingue_final = pd.crosstab(df_final['rango_puntaje'], df_final['cole_bilingue'])
tabla_porcentajes_puntaje_colegio_bilingue_final = tabla_porcentajes_puntaje_colegio_bilingue_final.div(tabla_porcentajes_puntaje_colegio_bilingue_final.sum(axis=1), axis=0).round(3) * 100
```

```python
# Unir proporciones originales y finales en un solo DataFrame
comparacion_proporciones = pd.DataFrame({
    'Original_S': tabla_porcentajes_puntaje_colegio_bilingue['S'],
    'Original_N': tabla_porcentajes_puntaje_colegio_bilingue['N'],
    'Final_S': tabla_porcentajes_puntaje_colegio_bilingue_final['S'],
    'Final_N': tabla_porcentajes_puntaje_colegio_bilingue_final['N'],
})

# Calcular diferencia en puntos porcentuales
comparacion_proporciones['Diferencia_S'] = (comparacion_proporciones['Final_S'] - comparacion_proporciones['Original_S']).round(2)
comparacion_proporciones['Diferencia_N'] = (comparacion_proporciones['Final_N'] - comparacion_proporciones['Original_N']).round(2)

# Mostrar resultado
print(comparacion_proporciones)
```

```python
import pandas as pd
from itertools import product

# Filtrar solo colegios oficiales
df_colegio_oficial = df_copia_datos1[df_copia_datos1['cole_naturaleza'] == 'OFICIAL']

# Variables a analizar
cols = ['fami_tienecomputador', 'fami_tieneinternet', 'fami_tieneautomovil', 'fami_tienelavadora']

# Subset con columnas relevantes
df_permutaciones = df_colegio_oficial[cols + ['desemp_ingles', 'fami_estratovivienda']]

# Contar la cantidad de veces que cada combinacion de las variables a analizar esta presente en el DataFrame y calcular su porcentaje 
conteo_combinaciones = df_permutaciones.value_counts(subset=cols).reset_index()
conteo_combinaciones.columns = cols + ['conteo']

total_conteo = conteo_combinaciones['conteo'].sum()
conteo_combinaciones['porcentaje'] = round((conteo_combinaciones['conteo'] / total_conteo) * 100, 2)

# Mostrar unicamente la columna de porcentaje 
porcentaje_df = conteo_combinaciones[cols + ['porcentaje']]

# Lista con frecuencia de desempeños en inglés
def frecuencia_desempeno(grupo):
    frec = grupo['desemp_ingles'].value_counts().sort_values(ascending=False)
    return ', '.join([f"{d}:{c}" for d, c in frec.items()])

desemp_freq = df_permutaciones.groupby(cols).apply(frecuencia_desempeno).reset_index(name='frecuencia_desemp_ingles')

# Lista con frecuencia de estratos 
def frecuencia_estratos(grupo):
    frec = grupo['fami_estratovivienda'].value_counts().sort_values(ascending=False)
    return ', '.join([f"{e}:{c}" for e, c in frec.items()])

estrato_freq = df_permutaciones.groupby(cols).apply(frecuencia_estratos).reset_index(name='frecuencia_estratos')

# Generar todas las combinaciones posibles de Si y No para las variables a analizar
combinaciones_posibles = list(product(['Si', 'No'], repeat=4))
df_combinaciones = pd.DataFrame(combinaciones_posibles, columns=cols)

# Unir los resultados
df_resultado_oficial = (
    df_combinaciones
    .merge(porcentaje_df, on=cols, how='left')
    .merge(desemp_freq, on=cols, how='left')
    .merge(estrato_freq, on=cols, how='left')
    .fillna({
        'porcentaje': 0,
        'frecuencia_desemp_ingles': 'Sin datos',
        'frecuencia_estratos': 'Sin datos'
    })
)

# Ordenar por porcentaje descendente 
df_resultado_oficial = df_resultado_oficial.sort_values(by='porcentaje', ascending=False).reset_index(drop=True)

# Mostrar resultado
print(df_resultado_oficial)
```
