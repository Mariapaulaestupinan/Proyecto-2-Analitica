A partir de los datos que tenemos, queremos identificar relaciones entre las columnas que contienen datos completos y aquellas que tienen datos faltantes, con el fin de poder estimar y completar los valores ausentes en lugar de eliminar las filas incompletas.

```python
import pandas as pd

# Crear los rangos de puntaje
rango_puntaje_global = [0, 20, 40, 60, 80, 100]
rango_labels = ['0-20', '20-40', '40-60', '60-80', '80-100']
df['rango_puntaje'] = pd.cut(df['punt_ingles'], bins=rango_puntaje_global, labels=rango_labels, right=False)

# Tabla cruzada: cuenta cuántos "S" y "N" hay por cada rango
tabla_cruzada = pd.crosstab(df['rango_puntaje'], df['cole_bilingue'])

# Convertimos a porcentaje sobre el total por fila (por rango)
tabla_porcentajes_puntaje_colegio_bilingue = tabla_cruzada.div(tabla_cruzada.sum(axis=1), axis=0).round(4) * 100

# Mostrar la tabla final
print(tabla_porcentajes_puntaje_colegio_bilingue)
```

```python
#Al DataFrame se le agrega una columna llamada rango_puntaje, que asigna un rango basado en el puntaje de inglés.
rango_puntaje_global_ajustado = [-0.01] + rango_puntaje_global[1:]

df_copia_datos['rango_puntaje'] = pd.cut(
    df_copia_datos['punt_ingles'],
    bins=rango_puntaje_global_ajustado,
    labels=rango_labels,
    right=True  
)
```

```python
def asignar_cole_bilingue_segun_puntaje(df, tabla_porcentajes, col_rango='rango_puntaje', col_obj='cole_bilingue', margen=5):
    # Crear una copia del DataFrame original para no modificarlo directamente
    df = df.copy()

    # Obtener la lista de categorías (valores posibles en la columna objetivo)
    categorias = tabla_porcentajes.columns.tolist()  
    
    # Iterar sobre cada rango de puntaje definido en el índice de la tabla de porcentajes
    for rango in tabla_porcentajes.index:
        # Guardar en un diccionario el porcentaje objetivo para cada categoría en este rango
        porcentaje_obj = tabla_porcentajes.loc[rango].to_dict()

        # Filtrar el DataFrame para quedarnos solo con las filas que pertenecen a este rango
        mask_rango = df[col_rango] == rango
        df_rango = df.loc[mask_rango]

        # Contar cuántas veces aparece cada categoría en la columna objetivo dentro de este rango
        conteo_actual = df_rango[col_obj].value_counts(dropna=True).to_dict()

        # Asegurar que todas las categorías estén presentes en el conteo, aunque sea con valor 0
        for cat in categorias:
            if cat not in conteo_actual:
                conteo_actual[cat] = 0

        # Obtener los índices de las filas donde la columna objetivo está vacía (NaN) para asignarles categoría
        idx_nan = df_rango[df_rango[col_obj].isna()].index.tolist()

        # Para cada fila con valor NaN en la columna objetivo, intentar asignar una categoría adecuada
        for idx in idx_nan:
            asignado = False

            # Ordenar las categorías según el porcentaje objetivo (de mayor a menor)
            orden_categorias = sorted(categorias, key=lambda x: porcentaje_obj[x], reverse=True)

            # Probar asignar cada categoría en orden hasta encontrar una que no supere el margen permitido de desviación
            for cat in orden_categorias:
                # Crear una copia temporal del conteo actual y sumar uno a la categoría que se está evaluando
                conteo_temp = conteo_actual.copy()
                conteo_temp[cat] += 1

                # Calcular el nuevo porcentaje temporal de cada categoría después de la asignación hipotética
                total_temp = sum(conteo_temp.values())
                porcentaje_temp = {k: (v / total_temp) * 100 for k, v in conteo_temp.items()}

                # Verificar si la desviación entre los porcentajes actuales y objetivos está dentro del margen permitido para todas las categorías
                desviacion_ok = all(abs(porcentaje_temp.get(k, 0) - porcentaje_obj[k]) <= margen for k in porcentaje_obj)

                # Si la desviación está dentro del margen, asignar esta categoría y actualizar el conteo actual
                if desviacion_ok:
                    df.at[idx, col_obj] = cat
                    conteo_actual = conteo_temp
                    asignado = True
                    break  # Salir del ciclo de categorías, porque ya se asignó una categoría válida

            # Si no se pudo asignar ninguna categoría que cumpla con el margen, asignar la categoría con mayor porcentaje objetivo por defecto
            if not asignado:
                cat = orden_categorias[0]
                df.at[idx, col_obj] = cat
                conteo_actual[cat] += 1

    # Retornar el DataFrame modificado con las asignaciones hechas
    return df
```

```python
#Apliar la función
df_final = asignar_cole_bilingue_segun_puntaje(df_copia_datos, tabla_porcentajes_puntaje_colegio_bilingue)
df_final
```

```python
#Validar en el DataFrame final la proporción de cada valor de la columna cole_bilingue dentro de cada rango de puntaje.
tabla_porcentajes_puntaje_colegio_bilingue_final = pd.crosstab(df_final['rango_puntaje'], df_final['cole_bilingue'])
tabla_porcentajes_puntaje_colegio_bilingue_final = tabla_porcentajes_puntaje_colegio_bilingue_final.div(tabla_porcentajes_puntaje_colegio_bilingue_final.sum(axis=1), axis=0).round(3) * 100
```

```python
# Unir proporciones originales y finales en un solo DataFrame
comparacion_proporciones = pd.DataFrame({
    'Original_S': tabla_porcentajes_puntaje_colegio_bilingue['S'],
    'Original_N': tabla_porcentajes_puntaje_colegio_bilingue['N'],
    'Final_S': tabla_porcentajes_puntaje_colegio_bilingue_final['S'],
    'Final_N': tabla_porcentajes_puntaje_colegio_bilingue_final['N'],
})

# Calcular diferencia en puntos porcentuales
comparacion_proporciones['Diferencia_S'] = (comparacion_proporciones['Final_S'] - comparacion_proporciones['Original_S']).round(2)
comparacion_proporciones['Diferencia_N'] = (comparacion_proporciones['Final_N'] - comparacion_proporciones['Original_N']).round(2)

# Mostrar resultado
print(comparacion_proporciones)
```

```python
import pandas as pd
from itertools import product

# Filtrar solo colegios oficiales
df_colegio_oficial = df_copia_datos1[df_copia_datos1['cole_naturaleza'] == 'OFICIAL']

# Variables a analizar
cols = ['fami_tienecomputador', 'fami_tieneinternet', 'fami_tieneautomovil', 'fami_tienelavadora']

# Subset con columnas relevantes
df_permutaciones = df_colegio_oficial[cols + ['desemp_ingles', 'fami_estratovivienda']]

# Contar la cantidad de veces que cada combinacion de las variables a analizar esta presente en el DataFrame y calcular su porcentaje 
conteo_combinaciones = df_permutaciones.value_counts(subset=cols).reset_index()
conteo_combinaciones.columns = cols + ['conteo']

total_conteo = conteo_combinaciones['conteo'].sum()
conteo_combinaciones['porcentaje'] = round((conteo_combinaciones['conteo'] / total_conteo) * 100, 2)

# Mostrar unicamente la columna de porcentaje 
porcentaje_df = conteo_combinaciones[cols + ['porcentaje']]

# Lista con frecuencia de desempeños en inglés
def frecuencia_desempeno(grupo):
    frec = grupo['desemp_ingles'].value_counts().sort_values(ascending=False)
    return ', '.join([f"{d}:{c}" for d, c in frec.items()])

desemp_freq = df_permutaciones.groupby(cols).apply(frecuencia_desempeno).reset_index(name='frecuencia_desemp_ingles')

# Lista con frecuencia de estratos 
def frecuencia_estratos(grupo):
    frec = grupo['fami_estratovivienda'].value_counts().sort_values(ascending=False)
    return ', '.join([f"{e}:{c}" for e, c in frec.items()])

estrato_freq = df_permutaciones.groupby(cols).apply(frecuencia_estratos).reset_index(name='frecuencia_estratos')

# Generar todas las combinaciones posibles de Si y No para las variables a analizar
combinaciones_posibles = list(product(['Si', 'No'], repeat=4))
df_combinaciones = pd.DataFrame(combinaciones_posibles, columns=cols)

# Unir los resultados
df_resultado_oficial = (
    df_combinaciones
    .merge(porcentaje_df, on=cols, how='left')
    .merge(desemp_freq, on=cols, how='left')
    .merge(estrato_freq, on=cols, how='left')
    .fillna({
        'porcentaje': 0,
        'frecuencia_desemp_ingles': 'Sin datos',
        'frecuencia_estratos': 'Sin datos'
    })
)

# Ordenar por porcentaje descendente 
df_resultado_oficial = df_resultado_oficial.sort_values(by='porcentaje', ascending=False).reset_index(drop=True)

# Mostrar resultado
print(df_resultado_oficial)
```

```python
# Función para convertir un string con formato 'clave: valor, clave: valor' en un diccionario {clave: int(valor)}
def parse_estratos(s):
    pares = s.split(',')  # Separar el string por comas para obtener cada par clave:valor
    d = {}
    for par in pares:
        par = par.strip()  # Eliminar espacios en blanco alrededor del par clave:valor
        if ':' in par:
            clave, valor = par.split(':')  # Separar clave y valor por los dos puntos
            d[clave.strip()] = int(valor.strip())  # Guardar en el diccionario la clave con valor entero
    return d

# Aplicar la función anterior a la columna 'frecuencia_estratos' para crear una nueva columna con diccionarios
df_resultado_oficial['estratos_dict'] = df_resultado_oficial['frecuencia_estratos'].apply(parse_estratos)

# Función para calcular el porcentaje que representa cada valor respecto al total de su diccionario
def dict_to_pct(d):
    total = sum(d.values())  # Sumar todos los valores del diccionario para obtener el total
    return {k: (v / total) * 100 for k, v in d.items()}  # Calcular el porcentaje de cada valor

# Aplicar la función de porcentaje a la columna de diccionarios y expandir el resultado en columnas separadas
df_pct = df_resultado_oficial['estratos_dict'].apply(dict_to_pct).apply(pd.Series).fillna(0)

# Unir el DataFrame original (sin las columnas originales de strings y diccionarios) con el DataFrame de porcentajes
df_resultado_oficial_estratos = pd.concat(
    [df_resultado_oficial.drop(columns=['frecuencia_estratos', 'estratos_dict']), df_pct], axis=1
)

# Mostrar el DataFrame resultante con las columnas de porcentajes por estrato
print(df_resultado_oficial_estratos)
```

```python
def asignar_estrato_oficial(grupo):
    # Obtener la combinación actual del grupo (nombre del grupo, que es una tupla)
    combo = grupo.name  
    # Total de filas en este grupo
    total = len(grupo)
    
    # Filtrar la fila correspondiente en df_resultado_oficial_estratos
    # Según la combinación exacta de las cuatro variables indicadas
    filtro = (
        (df_resultado_oficial_estratos['fami_tienecomputador'] == combo[0]) &
        (df_resultado_oficial_estratos['fami_tieneinternet'] == combo[1]) &
        (df_resultado_oficial_estratos['fami_tieneautomovil'] == combo[2]) &
        (df_resultado_oficial_estratos['fami_tienelavadora'] == combo[3])
    )
    
    # Extraer la fila con los porcentajes de estratos para esta combinación
    fila_estratos = df_resultado_oficial_estratos.loc[filtro]
    if fila_estratos.empty:
        # Si no se encuentra la combinación, retornar el grupo sin cambios
        return grupo  

    # Obtener los porcentajes objetivo de estratos de la fila encontrada,
    # seleccionando solo columnas que contienen la palabra 'Estrato' y convirtiendo a float
    porcentajes_obj = fila_estratos.iloc[0].filter(like='Estrato').astype(float)
    estratos_disponibles = porcentajes_obj.index.tolist()

    # Separar las filas que ya tienen estrato asignado de las que no
    asignados = grupo[grupo['fami_estratovivienda'].notna()]
    sin_asignar = grupo[grupo['fami_estratovivienda'].isna()]
    n_sin_asignar = len(sin_asignar)

    # Calcular el conteo actual de estratos ya asignados y convertir a porcentaje del total
    conteo_actual = asignados['fami_estratovivienda'].value_counts()
    porcentaje_actual = (conteo_actual / total) * 100
    
    # Calcular cuánto porcentaje falta asignar para cada estrato,
    # rellenando con 0 si algún estrato no está presente en conteo_actual
    porcentaje_faltante = porcentajes_obj - porcentaje_actual.reindex(porcentajes_obj.index, fill_value=0)
    porcentaje_faltante = porcentaje_faltante.clip(lower=0)  # No permitir valores negativos

    # Calcular cuántas filas hay que asignar a cada estrato según el porcentaje faltante
    filas_por_asignar = (porcentaje_faltante * total / 100).round().astype(int)
    # Ajustar la diferencia en caso de que la suma no coincida con el número de filas por asignar
    diferencia = n_sin_asignar - filas_por_asignar.sum()

    if diferencia != 0:
        # Ajustar la asignación sumando o restando a la categoría con mayor porcentaje faltante
        estrato_ajuste = filas_por_asignar.idxmax()
        filas_por_asignar[estrato_ajuste] += diferencia

    # Construir la lista de asignaciones con la cantidad determinada para cada estrato
    asignaciones = []
    for estrato, cantidad in filas_por_asignar.items():
        asignaciones.extend([estrato] * cantidad)

    # Ajustar en caso de que la cantidad asignada no coincida exactamente con n_sin_asignar
    if len(asignaciones) > n_sin_asignar:
        asignaciones = asignaciones[:n_sin_asignar]  # Cortar exceso
    elif len(asignaciones) < n_sin_asignar:
        estrato_max = porcentajes_obj.idxmax()
        asignaciones.extend([estrato_max] * (n_sin_asignar - len(asignaciones)))  # Completar con el estrato mayoritario

    # Mezclar aleatoriamente las asignaciones para evitar sesgo en el orden
    np.random.shuffle(asignaciones)

    # Asignar los estratos calculados a las filas que no tenían estrato asignado
    grupo.loc[grupo['fami_estratovivienda'].isna(), 'fami_estratovivienda'] = asignaciones

    # Validar el error entre la distribución final y la distribución objetivo
    conteo_final = grupo['fami_estratovivienda'].value_counts(normalize=True) * 100
    error = (conteo_final.reindex(porcentajes_obj.index, fill_value=0) - porcentajes_obj).abs()

    if error.max() > 5:
        # Si el error máximo es mayor a 5%, reasignar todo al segundo estrato mayoritario
        segundo_estrato = porcentajes_obj.sort_values(ascending=False).index[1]
        grupo['fami_estratovivienda'] = segundo_estrato

        print(f"Advertencia: Error > 5% para combinación {combo}. Se reasignó todo a {segundo_estrato}.")

    # Retornar el grupo con las asignaciones realizadas
    return grupo
```

```python
def calcular_participacion_por_combinacion(df):

    # Definir las columnas que se usarán para formar las combinaciones
    cols = ['fami_tienecomputador', 'fami_tieneinternet', 'fami_tienelavadora', 'fami_tieneautomovil']

    # Crear una nueva columna llamada 'combinacion' que agrupa los valores de las columnas definidas en una tupla por fila
    df['combinacion'] = df[cols].apply(lambda row: tuple(row.values), axis=1)

    # Agrupar el DataFrame por 'combinacion' y 'fami_estratovivienda' para contar cuántas filas hay en cada grupo
    conteo = df.groupby(['combinacion', 'fami_estratovivienda']).size().reset_index(name='cuenta')

    # Calcular el total de filas por cada combinación para poder calcular porcentajes dentro de cada grupo de combinación
    total_por_comb = conteo.groupby('combinacion')['cuenta'].transform('sum')

    # Calcular el porcentaje que representa cada estrato dentro de su combinación correspondiente
    conteo['porcentaje_participacion'] = (conteo['cuenta'] / total_por_comb) * 100

    # Devolver el DataFrame con la combinación, estrato, cuenta y porcentaje de participación
    return conteo
```

```python
def imputar_educacion_madre(df, tabla_porcentajes):
    # Crear una copia del dataframe para no modificar el original
    df_copy = df.copy()

    # Iterar sobre cada estrato en la tabla de porcentajes objetivo
    for estrato in tabla_porcentajes.index:
        # Crear una máscara para filtrar las filas que pertenecen a este estrato
        mask_estrato = df_copy['fami_estratovivienda'] == estrato
        sub_df = df_copy.loc[mask_estrato]

        # Si no hay valores nulos en 'fami_educacionmadre' para este estrato, continuar con el siguiente estrato
        if sub_df['fami_educacionmadre'].isna().sum() == 0:
            continue

        # Obtener la distribución objetivo de educación de madre para este estrato desde la tabla de porcentajes
        dist_objetivo = tabla_porcentajes.loc[estrato]

        # Contar la cantidad actual de cada nivel de educación en el subconjunto (excluyendo valores nulos)
        conteo_actual = sub_df['fami_educacionmadre'].value_counts()
        total_actual = conteo_actual.sum()  # Total de registros no nulos en este estrato

        # Asegurar que el conteo actual incluya todos los niveles de educación (llenar con 0 si falta alguno)
        conteo_actual = conteo_actual.reindex(dist_objetivo.index, fill_value=0)

        # Obtener los índices de las filas donde la educación de la madre está nula para este estrato
        indices_nulos = df_copy.loc[mask_estrato & df_copy['fami_educacionmadre'].isna()].index

        # Iterar sobre cada índice de fila con valor nulo para imputar
        for idx in indices_nulos:
            # Ordenar los niveles de educación por su porcentaje objetivo, de mayor a menor
            candidatos = dist_objetivo.sort_values(ascending=False).index.tolist()

            asignado = None  # Variable para guardar la educación que se asignará

            # Probar asignar cada nivel de educación en orden para minimizar desviación
            for educ in candidatos:
                # Simular la asignación sumando uno al conteo actual para ese nivel de educación
                conteo_sim = conteo_actual.copy()
                conteo_sim[educ] += 1
                total_sim = total_actual + 1  # Nuevo total después de la asignación
                porcentaje_sim = (conteo_sim / total_sim) * 100  # Calcular el nuevo porcentaje simulado

                # Calcular la diferencia absoluta entre la simulación y la distribución objetivo
                diferencia = (porcentaje_sim - dist_objetivo).abs()

                # Si la diferencia máxima es menor o igual a 5%, se acepta la asignación
                if diferencia.max() <= 5:
                    asignado = educ
                    break  # Salir del ciclo, asignación aceptable encontrada

            # Si ningún candidato cumple la condición de error <= 5%
            if asignado is None:
                # Si hay más de un candidato, asignar el segundo mejor porcentaje
                if len(candidatos) > 1:
                    asignado = candidatos[1]
                else:
                    # Si solo hay un candidato, asignar ese
                    asignado = candidatos[0]

            # Asignar el valor imputado al dataframe
            df_copy.at[idx, 'fami_educacionmadre'] = asignado

            # Actualizar conteo real para futuras asignaciones
            conteo_actual[asignado] += 1
            total_actual += 1

    # Devolver el dataframe con las imputaciones realizadas
    return df_copy
```

```python
def imputar_educacion_padre(df, tabla_porcentajes):
    # Crear una copia del dataframe para no modificar el original directamente
    df_copy = df.copy()

    # Iterar sobre cada estrato en la tabla de porcentajes objetivo
    for estrato in tabla_porcentajes.index:
        # Filtrar las filas del dataframe que corresponden al estrato actual
        mask_estrato = df_copy['fami_estratovivienda'] == estrato
        sub_df = df_copy.loc[mask_estrato]

        # Si en este estrato no hay valores nulos en 'fami_educacionpadre', pasar al siguiente estrato
        if sub_df['fami_educacionpadre'].isna().sum() == 0:
            continue

        # Obtener la distribución porcentual objetivo para este estrato desde la tabla de porcentajes
        dist_objetivo = tabla_porcentajes.loc[estrato]

        # Contar la cantidad actual de cada nivel de educación del padre en el subgrupo sin valores nulos
        conteo_actual = sub_df['fami_educacionpadre'].value_counts()
        total_actual = conteo_actual.sum()

        # Asegurar que todos los niveles de educación estén presentes en el conteo, rellenando con 0 donde falten
        conteo_actual = conteo_actual.reindex(dist_objetivo.index, fill_value=0)

        # Obtener los índices de las filas que necesitan imputación (valores nulos en 'fami_educacionpadre' dentro del estrato)
        indices_nulos = df_copy.loc[mask_estrato & df_copy['fami_educacionpadre'].isna()].index

        # Iterar sobre cada fila que requiere imputación
        for idx in indices_nulos:
            # Ordenar los niveles educativos candidatos según su porcentaje objetivo, de mayor a menor
            candidatos = dist_objetivo.sort_values(ascending=False).index.tolist()

            asignado = None

            # Intentar asignar cada candidato verificando que la desviación máxima no supere el 5%
            for educ in candidatos:
                # Simular la asignación de este nivel educativo incrementando el conteo actual
                conteo_sim = conteo_actual.copy()
                conteo_sim[educ] += 1
                total_sim = total_actual + 1

                # Calcular el porcentaje simulado tras la asignación
                porcentaje_sim = (conteo_sim / total_sim) * 100

                # Calcular la desviación absoluta entre el porcentaje simulado y el objetivo
                diferencia = (porcentaje_sim - dist_objetivo).abs()

                # Si la desviación máxima está dentro del rango aceptable, asignar este nivel educativo y salir del ciclo
                if diferencia.max() <= 5:
                    asignado = educ
                    break  # Asignación aceptable encontrada

            # Si ningún candidato cumple con la condición de desviación, asignar el segundo mejor candidato si existe
            if asignado is None:
                if len(candidatos) > 1:
                    asignado = candidatos[1]
                else:
                    # Si solo hay un candidato, asignarlo directamente
                    asignado = candidatos[0]

            # Asignar el nivel educativo elegido al dataframe copiado en el índice correspondiente
            df_copy.at[idx, 'fami_educacionpadre'] = asignado

            # Actualizar el conteo actual y el total para reflejar esta nueva asignación
            conteo_actual[asignado] += 1
            total_actual += 1

    # Retornar el dataframe con las imputaciones realizadas
    return df_copy
```

```python
# Aplicar la función de imputación para la educación del padre en el dataframe df_final3,
# usando la distribución porcentual por estrato contenida en tabla_porcentajes_estrato_educacion_padre.
df_final4 = imputar_educacion_padre(df_final3, tabla_porcentajes_estrato_educacion_padre)
# Mostrar el dataframe resultante con los valores imputados
df_final4
```

```python
# Crear una tabla cruzada entre el estrato de vivienda y la educación del padre en df_final4
evaluacion4 = pd.crosstab(df_final4['fami_estratovivienda'], df_final4['fami_educacionpadre'])

# Calcular la suma total de cada fila (es decir, total de registros por estrato)
suma_por_fila = evaluacion4.sum(axis=1)

# Convertir los conteos absolutos a porcentajes dentro de cada estrato
# Dividiendo cada celda por la suma total de su fila y multiplicando por 100 para obtener %
evaluacion4 = evaluacion4.div(suma_por_fila, axis=0) * 100

# Mostrar la tabla con los porcentajes de distribución de la educación del padre por estrato
print(evaluacion4)
```
